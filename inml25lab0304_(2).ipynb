**Plottare pdf di una legge normale μ=100, σ=15**
- dist = norm(loc=μ, scale=σ): Questa riga crea un oggetto dist che rappresenta la distribuzione normale utilizzando la funzione norm da scipy.stats
- x = np.linspace(40, 160, 101): Questa riga crea un array x di 101 valori equidistanti tra 40 e 160 usando la funzione linspace da numpy
- y = dist.pdf(x): Questa riga calcola la PDF della distribuzione normale (dist)
- plt.plot(x, y, label=f'pdf N({μ}, {σ}²)'): Questa riga crea un grafico a linee usando matplotlib.pyplot. I valori in x sono usati per l'asse orizzontale, i valori in y per l'asse verticale. L'etichetta del grafico è impostata come "pdf N(μ, σ²)" usando una f-string per formattare i valori di μ e σ.

distribuzione normale è N(μ, σ²), dove:
N indica la distribuzione normale
μ rappresenta la media
σ² rappresenta la varianza

```python
μ = 100
σ = 15

dist = norm(loc=μ, scale=σ)

x = np.linspace(40, 160, 101)

y = dist.pdf(x)

plt.plot(x, y, label=f'pdf N({μ}, {σ}²)')
plt.legend()

plt.show()
```

```python
x
```

**Plottare cdf di una legge normale μ=100, σ=15**


```python
z = dist.cdf(x)

plt.plot(x, z, label=f'CdF N({μ}, {σ}²)')
plt.legend()

plt.show()
```

**Generare 2000 vv.aa. con questa distribuzione**

- sample = norm.rvs(loc=μ, scale=σ, size=n): Questa riga genera il campione di numeri casuali.

```python
n = 2000

sample = norm.rvs(loc=μ, scale=σ, size=n)
sample
```

```python
from scipy.stats import beta
```

```python
# Genero 100000 massimi di 2000 uniformi su [0,1]

N = 100000
n = 2000

U_max = beta(a=n, b=1).rvs(size=N)
U_max
```

```python
# Genero 100000 massimi di 2000 Gaussiane come sopra

norm.ppf(U_max, loc=μ, scale=σ)
```

$X_i\sim\mathcal N(\mu,\sigma^2)$ indipendenti

$Z=\max_i X_i$

$X_i=F^{-1}(U_i)$ con $U_i\sim\text{unif}(0,1)$

$Z=\max_i F^{-1}(U_i)=F^{-1}(\max_i U_i)$ perché $F$ e $F^{-1}$ sono crescenti

$\max_i U_i\sim\text{beta}(n,1)$


```python
# C'è da migliorare la precisione, perché stiamo lavorando con numeri troppo vicini a 1

U_min = beta(a=1, b=n).rvs(size=N)
U_min
```

```python
Z_large = norm.isf(U_min, loc=μ, scale=σ)
Z_large
```

```python
plt.hist(Z_large, bins=np.arange(125, 200, 1), density=True, width=4)
plt.show()
```

```python
# Posso addirittura plottare la distribuzione esatta

x_large = np.linspace(125, 200, 101)
cdf_max = beta.cdf(norm.cdf(x_large, loc=μ, scale=σ), a=n, b=1)

plt.plot(x_large, cdf_max, label='theoretical CdF')
plt.ecdf(Z, label='empirical CdF')

plt.legend()
plt.show()
```

```python
pdf_max = beta.pdf(norm.cdf(x_large, loc=μ, scale=σ), a=n, b=1) * norm.pdf(x_large, loc=μ, scale=σ)

plt.plot(x_large, pdf_max, label='theoretical pdf')
plt.hist(Z, bins=np.arange(125, 200, 1), density=True, width=4, label='empirical pdf')

plt.legend()
plt.show()
```

1. Generare un processo di Poisson di parametro 1 e plottare $N_t$ per un po' di traiettorie

2. Studiare la distribuzione di $N_5$ e verificare che è plausibilmente di Poisson con media 5


```python
from scipy.stats import expon
```

```python
m = 100
β = 1 # la media delle esponenziali

T = expon.rvs(scale=β, size=m)
plt.plot(T, '.')
plt.show()
```

```python
S = np.cumsum(T)
S
```

```python
S.shape
```

```python
S.reshape(-1, 1).shape
```

```python
t = np.linspace(0, 20, 1001)
```

```python
t.reshape(1, -1).shape
```

```python
S.reshape(-1, 1) <= t.reshape(1, -1)
```

```python
t = np.linspace(0, 20, 1001)
Nt = np.sum(S.reshape(-1, 1) <= t.reshape(1, -1), axis=0)

plt.plot(t, Nt)
```

```python
k = 5
T = expon.rvs(scale=β, size=(m, k))
S = np.cumsum(T, axis=0)

t = np.linspace(0, 20, 1001)
Nt = np.sum(S.reshape(-1, k, 1) <= t.reshape(1, 1, -1), axis=0)

for i in range(k):
  plt.plot(t, Nt[i], label=f'N_t^({i})')

plt.legend()
plt.show()
```

```python
N = 10000
T = expon.rvs(scale=β, size=(m, N))
S = np.cumsum(T, axis=0)
t = 5

Nt = np.sum(S <= t, axis=0)
Nt.shape
```

```python


plt.hist(Nt, bins=np.arange(0, 20, 1), density=True, width=0.9)
plt.show()
```
